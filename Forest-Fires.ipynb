```python
from math import sqrt

def correlation_coefficient(x, y):
    r = (sum((x - np.mean(x)) * (y - np.mean(y))) /
         (sqrt(sum((x - np.mean(x))**2)) * sqrt(sum((y - np.mean(y))**2))))
    return r
```

```python
# Rs = [correlation_coefficient(i, df.area) for i in df.columns]

R01 = correlation_coefficient(df.X, df.area)
R02 = correlation_coefficient(df.Y, df.area)
R03 = correlation_coefficient(df.FFMC, df.area)
R04 = correlation_coefficient(df.DMC, df.area)
R05 = correlation_coefficient(df.DC, df.area)
R06 = correlation_coefficient(df.ISI, df.area)
R07 = correlation_coefficient(df.temp, df.area)
R08 = correlation_coefficient(df.RH, df.area)
R09 = correlation_coefficient(df.wind, df.area)
R10 = correlation_coefficient(df.rain, df.area)
```

```python
def sum_of_squares_residuals(x, y):
    SSx = (sum(x * x) - (((sum(x))**2) / len(x)))
    SSy = (sum(y * y) - (((sum(y))**2) / len(y)))
    SSxy = sum(x * y) - (1 / len(x)*(sum(x))*(sum(y)))
    return SSx, SSy, SSxy
```

```python
from plotly.subplots import make_subplots
from plotly.offline import *
```

```python
fig = make_subplots(cols=2,rows=2, specs=[[{'type':'scatter'}, {'type':'scatter'}], [{'type':'scatter'}, {'type':'scatter'}]], shared_yaxes=True)

fig.add_scatter(x=df.temp,
               y=df.area,
                mode='markers',
               name='temp',
               row=1,col=1)

fig.add_scatter(x=df.rain,
               y=df.area,
                mode='markers',
               name='rain',
               row=1,col=2)

fig.add_scatter(x=df.wind,
               y=df.area,
                mode='markers',
               name='wind',
               row=2,col=1)

fig.add_scatter(x=df.RH,
               y=df.area,
                mode='markers',
               name='RH',
               row=2,col=2)


fig.update_layout({'title':{'text':'Area of fire by conditions', 'font':{'size':32}}})

fig.update_yaxes({'title':{'text':'area'}})

fig.update_xaxes({'title':{'text':'temp'}}, row=1, col=1)
fig.update_xaxes({'title':{'text':'rain'}}, row=1, col=2)
fig.update_xaxes({'title':{'text':'wind'}}, row=2, col=1)
fig.update_xaxes({'title':{'text':'RH'}}, row=2, col=2)


iplot(fig)
```
